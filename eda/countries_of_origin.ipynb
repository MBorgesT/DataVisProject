```python
import pandas as pd

import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import plotly

pio.renderers.default = 'notebook'
```

```python
df = pd.read_csv('../treated_datasets/countries_of_origin.tsv', sep='\t')
df.head()
```

```python
pd.DataFrame(df.groupby('country_of_origin').count())
```

```python
plot_df = pd.DataFrame(df.groupby('country_of_origin').count()).reset_index()
plot_df = plot_df.rename(columns={'tconst': 'count'})
plot_df = plot_df.sort_values('count', ascending=False)
plot_df.head()
```

```python
px.histogram(plot_df[:20], x='country_of_origin', y='count')
```
